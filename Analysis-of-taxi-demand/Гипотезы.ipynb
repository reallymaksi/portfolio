```python
# Линейный график всех поездок по Москве
def plot_hourly_all_trips(dataframe):
    fig = go.Figure()

    fig.add_trace(go.Scatter(x=dataframe['hour'], y=dataframe['n_taxi_start'],
                             mode='lines+markers', name='Такси',
                             line=dict(color='blue')))

    fig.add_trace(go.Scatter(x=dataframe['hour'], y=dataframe['n_carsharing_start'],
                             mode='lines+markers', name='Каршеринг',
                             line=dict(color='red'), yaxis='y2'))

    fig.update_layout(
        title='Распределение поездок по часам (такси и каршеринг)',
        xaxis=dict(title='Час суток', dtick=1),
        yaxis=dict(title='Количество поездок (такси)', side='left'),
        yaxis2=dict(title='Количество поездок (каршеринг)',
                    overlaying='y', side='right'),
        legend=dict(x=0.5, y=-0.2, xanchor='center', orientation='h'),
        height=600,
        width=1000
    )

    return fig
```

```python
# Сгруппируем данные по номеру часа дня
main_df['hour'] = main_df['timestamp'].dt.hour
hour_pivot = main_df.pivot_table(index='hour', # группируем по hour
                    values=['n_taxi_start',
                            'n_carsharing_start'],
                    aggfunc={'n_taxi_start': 'sum', # суммарное количество начатых поездок в такси
                             'n_carsharing_start': 'sum', # суммарное количество начатых поездок в каршеринге
                            }).reset_index()

fig = plot_hourly_all_trips(hour_pivot.iloc[:13])

fig.show()
```

**Альтернативная гипотеза (H₁):**
 - Альтернативная гипотеза утверждает, что **медианное процентное изменение у такси значительно выше**, чем у каршеринга. Таким образом, делается предположение, что спрос на такси продолжает активно расти после 8:00, в то время как спрос на каршеринг либо стабилизируется, либо начинает снижаться.

**Метрика оценки:**
 - Для оценки гипотезы используется **процентное изменение** количества начатых поездок между 8:00 и 9:00 в каждом районе Москвы. **Расчёт процентного изменения количества поездок между 8:00 и 9:00**

Для анализа поведения пользователей такси и каршеринга в утренние часы были предприняты следующие шаги:

1. **Выбор необходимых признаков.**  
   Из датафрейма `main_df` были отобраны только нужные столбцы: количество поездок (`n_taxi_start`, `n_carsharing_start`), час начала поездки (`hour`), признак выходного/праздничного дня (`is_holiday_or_weekend`) и район города (`district_name`).

2. **Фильтрация по условиям.**  
   Оставлены только строки, где:
   - день рабочий (`is_holiday_or_weekend == 0`);
   - час равен 8 или 9 (`hour` ∈ [8, 9]).

3. **Агрегация данных.**  
   Для каждого района подсчитано суммарное количество поездок в 8:00 и в 9:00 отдельно для такси и каршеринга.

4. **Расчёт процентного изменения.**  
   Вычислено, насколько в процентах изменилось количество поездок между 8:00 и 9:00 утра по каждому району.

---

**Формула процентного изменения**

Формула расчёта процентного изменения количества поездок:

$$
\Delta\% = \left( \frac{N_9 - N_8}{N_8} \right) \times 100
$$

где:

- $N_8$ — количество начатых поездок в 8:00;
- $N_9$ — количество начатых поездок в 9:00;
- $\Delta\%$ — процентное изменение между 8 и 9 часами.

---

**Пояснение формулы**

- Если $\Delta\% > 0$, значит спрос вырос (в 9:00 поездок больше, чем в 8:00);
- Если $\Delta\% < 0$, значит спрос упал (в 9:00 поездок меньше, чем в 8:00);
- Если $\Delta\% = 0$, спрос остался прежним.

Этот расчёт позволяет сравнить динамику спроса в утренние часы **по районам Москвы** для такси и каршеринга.

```python
# Тест Шапиро-Уилка для такси
stat_taxi, pval_taxi = shapiro(delta_df['taxi_percent_change'].dropna())
print("Такси: p-value Шапиро-Уилка =", pval_taxi)
if pval_taxi < alpha:
    print("Распределение такси не нормально")
else:
    print("Распределение такси близко к нормальному")

# Тест Шапиро-Уилка для каршеринга
stat_car, pval_car = shapiro(delta_df['carsharing_percent_change'].dropna())
print("\nКаршеринг: p-value Шапиро-Уилка =", pval_car)
if pval_car < alpha:
    print("Распределение каршеринга не нормально")
else:
    print("Распределение каршеринга близко к нормальному")
```

`Убедились что распределение у такси ненормальное, поэтому используем тест Манна-Уитни и Уилкоксона для проверки гипотез.
Тест Манна-Уитни оценивает, есть ли различия в распределении двух выборок, но не предполагает, что данные в этих выборках следуют нормальному распределению. В отличие от параметрических тестов, таких как t-тест, который требует нормальности данных, тест Манна-Уитни может использоваться на любых данных, независимо от их распределения. Основная цель теста — это сравнение рангов (местоположений) значений в двух группах. Он не проверяет средние значения, как это делает t-тест, а больше фокусируется на том, есть ли различия в том, как распределяются данные между двумя выборками.`

##### Применение тестов для проверки гипотезы

`На прошлом шаге мы убедились, что распределение у такси ненормальное, поэтому используем тест Уилкоксона для проверки гипотез`

```python
# Непараметрический тест Манна-Уитни
stat, p = mannwhitneyu(delta_df['taxi_percent_change'], delta_df['carsharing_percent_change'])
print(f'p-value: {p}')
```

`Значение p очень маленькое, значит статистически значимые различия присутствуют и нулевая гипотеза отвергается данным тестом`

```python
# Тест Уилкоксона
stat, pval = wilcoxon(delta_df['taxi_percent_change'], delta_df['carsharing_percent_change'])

print("p-value:", pval)
if pval < 0.05:
    print("Различие в процентных изменениях значимо.")
else:
    print("Различие незначимо.")
```

`Тест Уилкоксона — это непараметрический тест, предназначенный для параметрических или связаных выборок. То есть его применяют, когда мы имеем два набора данных, которые тесно связаны между собой (например, измерения до и после какого-то воздействия или изменения).`

```python
# Односторонний тест Уилкоксона
stat, pval = wilcoxon(delta_df['taxi_percent_change'],
                      delta_df['carsharing_percent_change'],
                      alternative='greater')

print("Статистика Уилкоксона:", stat)
print("p-value:", pval)
if pval < 0.05:
    print("Процентное изменение для такси значимо больше, чем для каршеринга (отвергаем H0).")
else:
    print("Нет доказательств, что процентное изменение для такси больше (не отвергаем H0).")
```

### Гипотеза №2 - Временные ряды количества поездок в такси / каршеринге стационарны

#### Предпосылки

**Единичный корень и обоснование теста Дики-Фуллера в анализе транспортного спроса**

В рамках анализа повседневного спроса на такси и каршеринг в Москве по дням 2024 года важным этапом является определение **статистической устойчивости временного ряда**. Стабильность характеристик ряда (среднего, дисперсии, автокорреляции) во времени напрямую влияет на возможность его корректного моделирования и прогнозирования. Многие модели (например, ARIMA, SARIMA) предполагают **стационарность**, которая нарушается при наличии **тренда, сезонности или структурных сдвигов**. Для формальной проверки этого свойства применяется тест на наличие единичного корня — **тест Дики-Фуллера** (*Augmented Dickey-Fuller test*).

Пусть временной ряд $y_t$, отражающий суточное количество начатых поездок, моделируется в авторегрессионной форме порядка $p$:

$$
y_t = \sum_{i=1}^p a_i y_{t-i} + \varepsilon_t, \quad \varepsilon_t \sim \text{i.i.d. } (0, \sigma^2)
$$

С использованием оператора лага $L$, где $L y_t = y_{t-1}$, модель может быть переписана в виде:

$$
a(L) y_t = \left(1 - \sum_{i=1}^p a_i L^i\right) y_t = \varepsilon_t
$$

Рассматривается **характеристический полином**:

$$
a(z) = 1 - \sum_{i=1}^p a_i z^i
$$

Если все корни уравнения $a(z) = 0$ лежат **вне единичного круга** на комплексной плоскости (т.е. $|z_i| > 1$), то процесс является **стационарным**. Однако, если хотя бы один корень равен по модулю единице (например, $z = 1$), то процесс содержит **единичный корень** и является **нестационарным**. В случае наличия $k$ единичных корней полином $a(z)$ можно представить в виде:

$$
a(z) = b(z)(1 - z)^k
$$

Тогда модель принимает вид:

$$
a(L) y_t = b(L) (1 - L)^k y_t = b(L) \Delta^k y_t = \varepsilon_t
$$

Здесь $\Delta^k y_t$ обозначает конечные разности порядка $k$. Таким образом, ряд $y_t$ является интегрированным порядка $k$, то есть $y_t \sim I(k)$. В частности, если $k = 1$, ряд становится стационарным после одного дифференцирования: $\Delta y_t = y_t - y_{t-1}$.

Для транспортных данных (например, количества поездок по дням) это означает, что даже если наблюдаются колебания и тренды (например, рост числа поездок в зимний сезон, изменения в будни против выходных и т.д.), мы можем формально проверить, является ли ряд $I(1)$. Тест Дики-Фуллера проверяет следующую гипотезу:

* **Нулевая гипотеза (H₀):** ряд содержит единичный корень, то есть $y_t \sim I(1)$, нестационарен.
* **Альтернативная гипотеза (H₁):** ряд стационарен.

Применение теста Дики-Фуллера в данной задаче позволяет определить, следует ли предварительно трансформировать данные (например, через дифференцирование), прежде чем проводить сравнительный или прогнозный анализ спроса на такси и каршеринг.

```python
def plot_seasonal_decomposition(df, column, model='multiplicative', period=24, figsize=(20, 12)):
    """
    Выполняет декомпозицию временного ряда на основные компоненты и строит графики для визуального анализа.

    Параметры:
    - df: pandas DataFrame с индексом типа DatetimeIndex, содержащий временной ряд.
    - column: имя столбца в df, данные которого подлежат анализу (например, 'n_carsharing_start').
    - model: тип модели декомпозиции — 'multiplicative' (мультипликативная) или 'additive' (аддитивная).
    Выбор модели зависит от характера сезонных колебаний в данных.
    - period: период сезонности, например, 24 для часовых данных с суточным циклом.
    - figsize: кортеж, задающий размер графика (ширина, высота).

    Что делает функция:
    - Преобразует индекс датафрейма в формат datetime и сортирует данные по времени.
    - Заполняет пропуски методом прямого заполнения (forward fill).
    - Проводит декомпозицию временного ряда на следующие компоненты:
    - Наблюдаемые значения (исходный ряд).
    - Тренд — долгосрочная тенденция изменения.
    - Сезонность — периодические колебания с заданным периодом.
    - Остатки — шум и нерегулярные колебания, не объяснённые трендом и сезонностью.
    - Строит четыре графика, каждый из которых отображает одну из этих компонент, для удобного визуального анализа.

    Вывод:
    - Графическая визуализация декомпозиции временного ряда, позволяющая выявить тренды, сезонные циклы и нерегулярные колебания в данных.
    """

    df = df.copy()

    df.index = pd.to_datetime(df.index)
    df = df.sort_index()

    df[column] = df[column].ffill()

    # Декомпозиция временного ряда
    res = sm.tsa.seasonal_decompose(df[column], model=model, period=period)

    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=figsize, sharex=True)

    res.observed.plot(ax=ax1, color='blue', linewidth=1.5)
    ax1.set_title('Исходные данные', fontsize=14, pad=10)
    ax1.set_ylabel(column, fontsize=12)
    ax1.grid(True, linestyle='--', alpha=0.7)

    # График тренда
    res.trend.plot(ax=ax2, color='green', linewidth=1.5)
    ax2.set_title('Тренд', fontsize=14, pad=10)
    ax2.set_ylabel('Тренд', fontsize=12)
    ax2.grid(True, linestyle='--', alpha=0.7)

    # График сезонности
    res.seasonal.plot(ax=ax3, color='orange', linewidth=1.5)
    ax3.set_title('Сезонность', fontsize=14, pad=10)
    ax3.set_ylabel('Сезонность', fontsize=12)
    ax3.grid(True, linestyle='--', alpha=0.7)

    # График остатков
    ax4.scatter(res.resid.index, res.resid, color='red', s=30, alpha=0.6)
    ax4.set_title('Остатки', fontsize=14, pad=10)
    ax4.set_ylabel('Остатки', fontsize=12)
    ax4.grid(True, linestyle='--', alpha=0.7)

    date_formatter = DateFormatter('%Y-%m-%d')
    ax4.xaxis.set_major_formatter(date_formatter)
    plt.setp(ax4.get_xticklabels(), rotation=45, ha='right', fontsize=10)

    fig.suptitle(f'Декомпозиция временного ряда ({column})', fontsize=16, y=1.02)
    plt.tight_layout()

    plt.show()
```

* Тренд (trend) - плавное долгосрочное изменение значений ряда
* Сезонность (seasonality) - сезонные колебания ряда
* Шум (noise) - шумовая/случайная компонента ряда

```python
def plot_acf_pacf(dataframe, feature, lags=30, figsize=(14, 5),
                  title='Анализ автокорреляции временного ряда'):
    """
    Строит ACF и PACF графики для указанного признака из датафрейма.

    Параметры:
    ----------
    dataframe : pd.DataFrame
        Исходный датафрейм с временным рядом
    feature : str
        Название столбца с временным рядом
    lags : int, optional
        Количество лагов для анализа (по умолчанию 30)
    figsize : tuple, optional
        Размер фигуры (по умолчанию (14, 5))
    title : str, optional
        Общий заголовок графиков (по умолчанию 'Анализ автокорреляции временного ряда')
    """
    # Настройки стиля
    plt.rcParams["figure.dpi"] = 100
    plt.rcParams["font.size"] = 12

    # Создаем фигуру
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)

    # График ACF
    plot_acf(dataframe[feature].dropna(),
             lags=lags,
             alpha=0.05,
             ax=ax1,
             title=f'ACF: {feature}',
             color='#1f77b4')
    ax1.grid(True, linestyle=':', alpha=0.7)
    ax1.set_xlabel('Лаг', fontsize=12)
    ax1.set_ylabel('Корреляция', fontsize=12)

    # График PACF
    plot_pacf(dataframe[feature].dropna(),
              lags=lags,
              alpha=0.05,
              ax=ax2,
              title=f'PACF: {feature}',
              color='#ff7f0e')
    ax2.grid(True, linestyle=':', alpha=0.7)
    ax2.set_xlabel('Лаг', fontsize=12)
    ax2.set_ylabel('Частичная корреляция', fontsize=12)

    # Общие настройки
    plt.suptitle(title, y=1.05, fontsize=14)
    plt.tight_layout()
    plt.show()
```

```python
def plot_acf_pacf(dataframe, feature, lags=30, figsize=(14, 5),
                  title='Анализ автокорреляции временного ряда'):
    """
    Строит графики автокорреляционной функции (ACF) и частичной автокорреляционной функции (PACF) для заданного временного ряда.

    Параметры:
    ----------
    dataframe : pd.DataFrame
        Исходный датафрейм, содержащий временной ряд с индексом времени.
    feature : str
        Название столбца в dataframe, данные которого будут анализироваться.
    lags : int, optional
        Максимальное количество лагов (запаздываний) для расчёта автокорреляций (по умолчанию 30).
    figsize : tuple, optional
        Размер графика в формате (ширина, высота) в дюймах (по умолчанию (14, 5)).
    title : str, optional
        Общий заголовок для отображаемых графиков (по умолчанию 'Анализ автокорреляции временного ряда').

    Что делает функция:
    - Рассчитывает и визуализирует автокорреляцию (ACF) — зависимость текущих значений ряда от его предыдущих значений.
    - Рассчитывает и визуализирует частичную автокорреляцию (PACF) — влияние каждого лага на текущие значения с учётом корреляции предыдущих лагов.
    - Помогает определить порядок моделей временных рядов, таких как ARIMA, для дальнейшего моделирования и прогнозирования.

    Вывод:
    - Два графика — ACF и PACF, отображающие значения корреляций по лагам с доверительными интервалами. Эти графики позволяют визуально оценить структуру временного ряда и выбрать параметры моделей для дальнейшего анализа.
    """

    # Настройки стиля
    plt.rcParams["figure.dpi"] = 100
    plt.rcParams["font.size"] = 12

    # Создаем фигуру
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)

    # График ACF
    plot_acf(dataframe[feature].dropna(),
             lags=lags,
             alpha=0.05,
             ax=ax1,
             title=f'ACF: {feature}',
             color='#1f77b4')
    ax1.grid(True, linestyle=':', alpha=0.7)
    ax1.set_xlabel('Лаг', fontsize=12)
    ax1.set_ylabel('Корреляция', fontsize=12)

    # График PACF
    plot_pacf(dataframe[feature].dropna(),
              lags=lags,
              alpha=0.05,
              ax=ax2,
              title=f'PACF: {feature}',
              color='#ff7f0e')
    ax2.grid(True, linestyle=':', alpha=0.7)
    ax2.set_xlabel('Лаг', fontsize=12)
    ax2.set_ylabel('Частичная корреляция', fontsize=12)

    # Общие настройки
    plt.suptitle(title, y=1.05, fontsize=14)
    plt.tight_layout()
    plt.show()
```

#### Формулировка гипотезы

**Формулировка гипотезы:**

* Предполагается, что временные ряды количества начатых поездок в Москве по дням за 2024 год для такси (`n_taxi_start`) и каршеринга (`n_carsharing_start`) имеют разную структуру временной зависимости. Для корректного анализа и последующего моделирования важно установить, являются ли эти ряды **стационарными** — то есть сохраняют ли они стабильное среднее и дисперсию во времени.

**Нулевая гипотеза (H₀):**

* Согласно нулевой гипотезе, **временной ряд имеет единичный корень**, то есть **не является стационарным**. Это означает, что его статистические характеристики (среднее, дисперсия) изменяются со временем, что делает модельные прогнозы менее устойчивыми без предварительного преобразования (например, дифференцирования).

**Альтернативная гипотеза (H₁):**

* Альтернативная гипотеза утверждает, что **ряд стационарен**, то есть не содержит единичного корня. Это означает, что уровень временного ряда стабилен во времени, а колебания происходят вокруг устойчивого среднего значения с постоянной дисперсией.

**Метрика оценки:**

* Для проверки гипотезы используется **тест Дики-Фуллера (Augmented Dickey-Fuller test)**. Ключевым параметром является p-значение, отражающее вероятность того, что наблюдаемый временной ряд имеет единичный корень. При p-значении ниже принятого уровня значимости (обычно 0.05) нулевая гипотеза отвергается.

**Контекст эксперимента:**

* Анализ проводится по временным рядам `n_taxi_start` и `n_carsharing_start`, агрегированным **по дням** за 2024 год, включая как будни, так и выходные. Выбор именно **дневного уровня агрегации** обусловлен стремлением выявить устойчивые макропаттерны спроса, не искажённые краткосрочными флуктуациями, характерными для почасовых данных. Почасовые ряды демонстрируют высокую волатильность, связанную с пиковыми периодами, погодой и случайными событиями, что затрудняет интерпретацию результатов теста на стационарность. Суточные значения позволяют сгладить эти кратковременные колебания и сфокусироваться на более стабильных трендах и сезонных эффектах.

**Ожидаемый эффект:**

* Ожидается, что **оба ряда окажутся нестационарными**, поскольку спрос на такси и каршеринг подвержен ярко выраженной сезонности (в течение недели, месяца, года), а также трендам (например, рост популярности сервиса). Это указывает на необходимость трансформации данных перед использованием в предиктивных моделях (например, через логарифмирование и дифференцирование).

**Связь с задачей:**

* Проверка стационарности — ключевой этап при построении временных моделей (например, ARIMA, SARIMA, Prophet). Результаты теста позволяют определить, какие преобразования необходимо применить к рядам, чтобы обеспечить корректность последующего анализа и прогнозирования, а также повысить интерпретируемость моделей в задачах управления транспортной системой города.

#### Проверка гипотезы

```python
def run_adf_test(series: pd.Series, title: str = 'Time Series') -> dict:
    """
    Проводит тест Дики-Фуллера на стационарность временного ряда.

    Параметры:
        series (pd.Series): Временной ряд, индекс должен быть по датам.
        title (str): Название ряда (для вывода).

    Возвращает:
        dict: Словарь с результатами теста.
    """
    print(f'ADF Test for: {title}')
    print('-' * 40)

    series_clean = series.dropna()

    result = adfuller(series_clean, autolag='AIC')

    output = {
        'ADF Statistic': result[0],
        'p-value': result[1],
        'Number of Lags Used': result[2],
        'Number of Observations Used': result[3],
        'Critical Values': result[4],
        'Stationary (5% level)': result[1] < 0.05
    }

    for key, value in output.items():
        if key == 'Critical Values':
            print('Critical Values:')
            for level, cv in value.items():
                print(f'    {level}: {cv:.4f}')
        else:
            print(f'{key}: {value}')

    print('-' * 40)
    if output['Stationary (5% level)']:
        print(f'✅ {title} считается стационарным (p < 0.05)')
    else:
        print(f'❌ {title} НЕ считается стационарным (p ≥ 0.05)')

    return output
```

```python
# Проверка данных на нормальность распределения тестом Шапиро-Уилка
def normality_test(data):
    statistic, p_value = stats.shapiro(data)

    if p_value < alpha:
        return print(f'{p_value} < {alpha} -> данные не распределены нормально')
    else:
        return print('оснований отвергать нормальность нет')
```

```python
normality_test(trips_per_person_high_access)
normality_test(trips_per_person_medium_access)
normality_test(trips_per_person_low_access)
```

Проверка будем проводить с помощью теста Краскела–Уоллиса, так как распределения не являются нормальными, а выборки имеют разные размеры.

```python
# Применяем тест Краскела-Уоллиса
stat, p = kruskal(trips_per_person_high_access, trips_per_person_medium_access, trips_per_person_low_access)

if p < alpha:
    print(f'p-value={p}')
    print('Отвергаем нулевую гипотезу -> есть статистически значимые различия хотя бы между двумя группами')
else:
    print('Нет оснований отвергнуть нулевую гипотезу -> нет оснований полагать, что группы различаются')
```

**Посмотрим между какими группами есть статистически значимые различия:**
 Будем проводить тест Манна-Уитни с поправкой Бонферрони.

- **Нулевая гипотеза (H₀)** - Распределения количества поездок на одного жителя в двух сравниваемых группах одинаковы — статистически значимых различий нет.
- **Альтернативная гипотеза (H₁)** - Распределения количества поездок на одного жителя в двух сравниваемых группах различаются — в одной из групп медиана статистически значимо выше или ниже, чем в другой.

```python
u1, p1 = mannwhitneyu(trips_per_person_high_access, trips_per_person_medium_access)
u2, p2 = mannwhitneyu(trips_per_person_high_access, trips_per_person_low_access)
u3, p3 = mannwhitneyu(trips_per_person_medium_access, trips_per_person_low_access)

# Альфа с поправкой Бонферрони
bonfer_alpha = alpha / 3  # 3 сравнения

print(f"Высокая доступность vs Средняя доступность: p = {p1} {'Отвергаем нулевую гипотезу' if p1 < bonfer_alpha else 'Нет оснований отвергнуть нулевую гипотезу'}")
print(f"Высокая доступность vs Низкая доступность: p = {p2} {'Отвергаем нулевую гипотезу' if p2 < bonfer_alpha else 'Нет оснований отвергнуть нулевую гипотезу'}")
print(f"Средняя доступность vs Низкая доступность: p = {p3} {'Отвергаем нулевую гипотезу' if p3 < bonfer_alpha else 'Нет оснований отвергнуть нулевую гипотезу'}")
```

```python
# Направленный тест Манна-Уитни, чтобы понять где медиана выше
def compare_groups(group1, group2, name1, name2):

    stat, p = mannwhitneyu(group1, group2, alternative='greater')

    if p < bonfer_alpha:
      return print(f"{name1} vs {name2}: p = {p} → Отвергаем нулевую гипотезу: в группе '{name1}' средние значения выше, чем в '{name2}'")

    else:
      return print(f"{name1} vs {name2}: p = {p} → Нет оснований отвергнуть нулевую гипотезу")


compare_groups(trips_per_person_high_access, trips_per_person_medium_access, "Высокая доступность", "Средняя доступность")
compare_groups(trips_per_person_high_access, trips_per_person_low_access, "Высокая доступность", "Низкая доступность")
compare_groups(trips_per_person_medium_access, trips_per_person_low_access, "Средняя доступность", "Низкая доступность")
```

**Вывод:**

Была проверена гипотеза о влиянии доступности метро на количество поездок в такси по районам. С помощью непараметрического теста **Краскела–Уоллиса** (p-value = **0.0007**) выявлены статистически значимые различия в распределениях количества поездок между группами районов с разной доступностью метро. Это позволило отвергнуть нулевую гипотезу о равенстве распределений.

Для уточнения, между какими группами наблюдаются различия, был проведён тест **Манна–Уитни** с поправкой Бонферрони (α = 0.05 / 3 ≈ 0.0167). Результаты показали наличие значимых различий между:

	•	«Высокая доступность» vs «Низкая доступность» (p = **0.00056**)
	•	«Средняя доступность» vs «Низкая доступность» (p = **0.00885**)

При этом различий между группами «Высокая» и «Средняя» доступность метро обнаружено не было.

Дополнительно был проведён **односторонний тест Манна–Уитни** (alternative='greater'), который показал, что медианное количество поездок в группах с высокой и средней доступностью выше, чем в группе с низкой доступностью.

### Гипотеза №4 - Стоимость недвижимости влияет на спрос поездок в такси.

**Н₀ (нулевая гипотеза)**: Количество поездок на такси в районах, сгруппированных по уровню цен на недвижимость, статистически значимо не различается.

**Н₁ (альтернативная гипотеза)**: Количество поездок хотя бы в одной группе имеет статистически значимое различие.

Сделаем сводную таблицу, в которой каждому району будет соответствовать общее количество поездок за год (n_taxi_start), цена на недвижимость, а также категория по уровню цен (высокий,средний,низкий уровень).

```python
# Считаем суммарное число поездок и население по району
trips_by_district = main_df.groupby('district_name').agg({
    'n_taxi_start': 'sum',
    'population_district_2023': 'first',
    'price_m2': 'first'
}).reset_index()

# Делаем дополнительный столбец с разбивкой по уровню цен (высокие,средние,низкие)
trips_by_district['price_group'] = pd.qcut(trips_by_district['price_m2'], q=3, labels=['Low', 'Medium', 'High'])
```

```python
# Посмотрим на гистограммах, как распределено количество поездок по группам
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Строим графики
sns.histplot(trips_by_district[trips_by_district['price_group'] == 'Low']['n_taxi_start'], ax=axes[0], kde=True).set(title='Low')
sns.histplot(trips_by_district[trips_by_district['price_group'] == 'Medium']['n_taxi_start'], ax=axes[1], kde=True).set(title='Medium')
sns.histplot(trips_by_district[trips_by_district['price_group'] == 'High']['n_taxi_start'], ax=axes[2], kde=True).set(title='High')

# Названия осей
axes[0].set_xlabel('Число поездок')
axes[0].set_ylabel('Частота')
axes[1].set_xlabel('Число поездок')
axes[1].set_ylabel('Частота')
axes[2].set_xlabel('Число поездок')
axes[2].set_ylabel('Частота')

# Общее название
plt.suptitle('Распределение числа поездок в зависимости от цен на недвижимость', fontsize=16)

plt.tight_layout()
plt.subplots_adjust(top=0.85)
plt.show()
```

На графиках видно,что распределение числа поездок отличается от нормального. Для дополнительной проверки сделаем тест Шапиро-Уилка.

```python
for group in ['Low', 'Medium', 'High']:
    data = trips_by_district[trips_by_district['price_group'] == group]['n_taxi_start']
    stat, p_value = stats.shapiro(data)
    print(f'{group} group - p-value: {p_value:.4f} → {"Ненормальное" if p_value < 0.05 else "Нормальное"} распределение')
```

Тест подтвердил что распределения во всех группах отличаются от нормального, поэтому для проверки гипотезы будем использовать непараметрический тест Краскела–Уоллиса.

```python
from scipy.stats import kruskal

# Формируем список значений с количеством поездок для каждой из групп
groups = trips_by_district.groupby('price_group')['n_taxi_start'].apply(list)

# Тест Краскела–Уоллиса
stat, p = kruskal(*groups)

print(f"p-value: {p:.6f}")

if p < 0.05:
    print("Различия между группами статистически значимы →  Oтвергаем H₀.")
else:
    print("Различия между группами статистически незначимы → Нет оснований отвергнуть H₀.")
```

Для того, чтобы понять между какими именно группами существуют различия, проведем Тест Тьюки (Tukey HSD).

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Применяем тест Тьюки для многократных сравнений
tukey_result = pairwise_tukeyhsd(
    trips_by_district['n_taxi_start'],  # Значения для теста
    trips_by_district['price_group'],   # Группирующие переменные
    alpha=0.05                         # Уровень значимости
)

# Выводим результаты
print(tukey_result)
```

#### Результаты теста Тьюки

1. **High vs Low**: Различие статистически значимо (p < 0.001). В районах с высокими ценами на недвижимость поездок существенно меньше, чем в районах с низкими(разница: −1,561 тыс.).

2. **High vs Medium**: Различие также статистически значимо (p < 0.001). Количество поездок в районах с высокими ценами значительно меньше, чем в районах со средними ценами (разница: −1,430 тыс.).

3. **Low vs Medium**: Различие не является статистически значимым (p ≈ 0.91). Это означает, что число поездок в районах с низкими и средними ценами на жильё примерно одинаково.


Таким образом, проверка гипотезы о влиянии стоиомсти недвижимости на спрос в такси с использованием теста  Краскела–Уоллиса, показала статистически значимые различия между количеством поездок в разных группах, в частности:
 - В районах с высокими ценами на недвижимость поездок существенно меньше, чем в районах с низкими или средними ценами.
 -  Число поездок в районах с низкими и средними ценами на жильё примерно примерно одинаково.

### Гипотеза №5 - Количество станций метро влияет на их загруженность.



**H₀(нулевая гипотеза)**: Количество станций метро (number_of_stations) не влияет на их загруженность (station_load).

**H₁(альтернативная гипотеза)**: Чем меньше станций в районе, тем выше их средняя загруженность(обратная зависимость).

Обоснование выдвинутой гипотезы: чем меньше станций тем выше концентрация пассажиров на каждой станции, как пример: район с 1 станцией метро будет принимать большее количество людей, чем район с 10 станциями, поскольку последний распределит загруженность между всеми

Для начала посмотрим на график зависимости

```python
data = main_df.copy()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x='number_of_stations', y='station_load', data=data, scatter_kws={'alpha': 0.3}, line_kws={'color': 'red'})
plt.title("Зависимость загруженности метро от числа станций в районе")
plt.xlabel("Число станций метро")
plt.ylabel("Загруженность метро")
plt.show()
```

Мы работаем с двумя категориальными переменными, number_of_stations - вещественная дискретная, а station_load - категориальная порядковая. Проверим данные на нормальность с помощью теста Д’Агостино, мы не пользуемся тестом Шапиро-Уилка потому что у нас достатачно большая выборка для оценки нормальности первый вариант предпочтительнее

```python
#группируем данные по районам
data['station_load_mean_district'] = data.groupby('district_name')['station_load'].transform('mean').round().astype(int)
data['number_of_stations_mean_district'] = data.groupby('district_name')['number_of_stations'].transform('mean').round().astype(int)
```

```python
stat, p = normaltest(data['number_of_stations_mean_district'])
print('p-value:', p)
```

Данные не распределены нормально, поэтому будем пользоваться непараметрическими методами(загруженность станций мы не проверяем, поскольку переменная категориальная), в нашем случае подходит корреляционный тест Спирмена

Почему именно тест Спирмена:

1) Работает с порядковыми данными

2) Не требует нормального распределения

3) Проверяет растет/убывает ли одна из переменных при изменении другой

Какие еще были варианты:

Критерий Краскела-Уоллиса (для сравнения групп) менее точен здесь, так как у нас конкретная гипотеза о связи, а не о различии групп.

Пирсон не подходит, так как требует интервальных данных и нормальности данных

```python
# Корреляция Спирмена
rho, p_value = stats.spearmanr(data['number_of_stations_mean_district'], data['station_load_mean_district'])

print(f"Коэффициент Спирмена (ρ): {rho:.3f}")
print(f"p-value: {p_value:.3f}")
```

У нас получился отрицательный коээфициент, он указывает на обратную зависимость между количеством станций и их загруженностью(Гипотеза H0 отвергается), такой коэф указывает на умеренную корелляцию(возможно есть третий фактор, который мы не успели рассмотреть, но данное исследование дает возможность предположить, что гипотеза H1 имеет место быть)

### Гипотеза №6 - Количество станций  влияет на спрос поездок в такси.



**Нулевая гипотеза (H₀)**: Нет корреляции между количеством станций метро (number_of_stations) и количеством поездок на такси (n_taxi_start).

**Альтернативная гипотеза (H₁)**: Существует положительная корреляция между числом станций и поездками на такси.

Обоснование выдвинутой гипотезы: Чем больше станций метро в районе, тем выше его транспортная доступность и плотность населения, как пример: офисные районы с развитым метрополитеном == высокий спрос на такси


Поскольку обе переменные количественные, а гипотеза предполагает монотоннуюсвязь, подходят:

Корреляция Пирсона - если связь линейная и данные распределены нормально.

Корреляция Спирмена - если связь монотонная, но нелинейная или данные не нормальные.
